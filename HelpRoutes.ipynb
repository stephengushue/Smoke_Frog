```python
df = pd.read_csv(file, parse_dates=[1])
'''loads csv file into pandas dataframe'''
```

```python
df['NEW_ROUTE']=''
'''adds new column in dataframe to add day designation letter to front of route'''
```

```python
df.loc[df['DAY']=='Monday', "NEW_ROUTE"] = 'M' + df['ROUTE']
df.loc[df['DAY']=='Tuesday', "NEW_ROUTE"] = 'T' + df['ROUTE']
df.loc[df['DAY']=='Wednesday', "NEW_ROUTE"] = 'W' + df['ROUTE']
df.loc[df['DAY']=='Thursday', "NEW_ROUTE"] = 'H' + df['ROUTE']
df.loc[df['DAY']=='Friday', "NEW_ROUTE"] = 'F' + df['ROUTE']
df.loc[df['DAY']=='Saturday', "NEW_ROUTE"] = 'S' + df['ROUTE']
```

```python
df
```

```python
df.drop(columns=['DISCARD', 'NEW_TOUTE'])
'''drops unused fields'''
```

### Further cleaning and melting

Here is where the scripting gets interesting. The following cell block melts the wide format of the csv file with multiple HELP columns and creates new rows with the help route to each driver, for example:
    
        (AS IS)
        OPERATOR NAME     DATE      HELP ROUTE 1     HELP ROUTE 2
        STEVE GUSHUE      04/19/18  MB67             MB68
        
        (MELTED)
        OPERATOR NAME     DATE          VALUE (OF HELP ROUTE COLUMNS)
        STEVE GUSHUE      04/19/2018    MB67
        STEVE GUSHUE      04/19/2018    MB68
        


```python
df2 = pd.DataFrame(pd.melt(df, id_vars=['DATE', 'OPERATOR', 'NEW_ROUTE', 'DAY'], value_vars=['HELP_1', 'HELP_2', 'HELP_3', 'DOWN']))
```
